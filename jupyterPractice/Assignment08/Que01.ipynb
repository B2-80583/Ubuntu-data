```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./Churn.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# done scaling because some colums has very high mean 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[['EstimatedSalary']])
df['EstimatedSalary'] = scaler.transform(df[['EstimatedSalary']])
df.describe()
```

```python
df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder

# create an encoder object
encoder = LabelEncoder()

# fit the data
encoder.fit(df['Surname'])

# transform the Gender column
df['Surname'] = encoder.transform(df['Surname'])

encoder.fit(df['Geography'])

# transform the Gender column
df['Geography'] = encoder.transform(df['Geography'])

encoder.fit(df['Gender'])

# transform the Gender column
df['Gender'] = encoder.transform(df['Gender'])
```

```python
df.head(10)
```

```python
df.drop(['Surname', 'CustomerId', 'HasCrCard'], axis=1, inplace=True)
```

```python
x = df.drop(['Exited'], axis = 1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, random_state=12345)
```

```python
def build_dt_model():
    from sklearn.tree import DecisionTreeClassifier
    model = DecisionTreeClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
def build_rd_model():
    from sklearn.ensemble import RandomForestClassifier
    model = RandomForestClassifier(n_estimators=100)
    model.fit(x_train, y_train)
    return model
```

```python
def build_cb_model():
    from catboost import CatBoostClassifier
    model = CatBoostClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
def build_xgb_model():
    from xgboost import XGBClassifier
    model = XGBClassifier()
    model.fit(x_train, y_train)
    return model
```

```python
def evaluate_model(model):
    y_true = y_test
    y_pred = model.predict(x_test)
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    accuracy = f"{accuracy_score(y_true, y_pred):.2f}"
    precision = f"{precision_score(y_true, y_pred):.2f}"
    recall = f"{recall_score(y_true, y_pred):.2f}"
    f1 = f"{f1_score(y_true, y_pred):.2f}"

    return accuracy, precision, recall, f1
```

```python
evaluation_report = []

model_dt = build_dt_model()
evaluation_report.append({"model": "Decision Tree", "metrics": evaluate_model(model_dt)})

model_rd = build_rd_model()
evaluation_report.append({"model": "Random Forest", "metrics": evaluate_model(model_rd)})

model_cb = build_cb_model()
evaluation_report.append({"model": "CatBoost", "metrics": evaluate_model(model_cb)})

model_xgb = build_xgb_model()
evaluation_report.append({"model": "XGB", "metrics": evaluate_model(model_xgb)})
```

```python
print(f"{'Algorithm':<20} => 'Accuracy', 'precision', 'recall', 'f1'")
print()
for info in evaluation_report:
    print(f"{info['model']:<20} => {info['metrics']}")
```
