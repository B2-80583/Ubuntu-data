### data preprocessing

```python
# get independent variables
x = df.drop('Profit', axis = 1)

y = df['Profit']
```

### create a model using Linear Regression

```python
from sklearn.linear_model import LinearRegression

# create a model
model = LinearRegression()

# train the model
model.fit(x, y)
```

### reduce the independent variables to only one for visualisation 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)

pca.fit(x)
reduced_x = pca.transform(x)
```

##  create a new model using reduced independent variable


```python
from sklearn.linear_model import LinearRegression

# create a model
model = LinearRegression()

# train the model using reduced independent variables
model.fit(reduced_x, y)
```

### result visualisation

```python
plt.scatter(reduced_x, df['Profit'])

Y_predict = model.predict(reduced_x)
plt.plot(reduced_x, Y_predict, color="red")
plt.scatter(reduced_x, Y_predict, color="red", label="regression line")

```
