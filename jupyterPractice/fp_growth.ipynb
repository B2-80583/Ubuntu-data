# fp groth

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### get the data from data source

```python
dataset = [
     ['E', 'A', 'D', 'B'],
    ['D', 'A', 'C', 'E', 'B'],
    ['C', 'A', 'B', 'E'],
    ['B', 'A', 'D'],
    ['D'],
    ['D', 'B'],
    ['A', 'D', 'E'],
    ['B', 'C']
]
```

### preprocessing the data

```python
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()

encoder.fit(dataset)

transformed_dataset = encoder.transform(dataset)

df_transormed = pd.DataFrame(transformed_dataset, columns=encoder.columns_)
df_transormed

```

```python
from mlxtend.frequent_patterns import fpgrowth
results = fpgrowth(df_transormed, use_colnames=True)
results
```
